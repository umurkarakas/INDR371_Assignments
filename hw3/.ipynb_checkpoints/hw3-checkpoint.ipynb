## Question 1

```python
def calculate(Q, revenues, unit_cost, penalty_cost, lower_limit, p_accepted):
    prod_probability = np.asarray([binom.pmf(k=i, n=Q, p=p_accepted) for i in range(Q+1)])
    total_revenues = np.asarray([revenues[i] if i in list(revenues.keys()) else revenues[max(list(revenues.keys()))] if i > max(list(revenues.keys())) else 0 for i in range(Q+1)])
    total_prod_cost = np.asarray([unit_cost*Q for i in range(Q+1)])
    total_penalty_cost = np.asarray([penalty_cost if i<lower_limit else 0 for i in range(Q+1)])
    expected_profit = np.dot(prod_probability, total_revenues-total_prod_cost-total_penalty_cost)
    #print("Production probabilities:", prod_probability)
    #print("Revenues:", total_revenues)
    #print("Total production cost:", total_prod_cost)
    #print("Total penalty cost:", total_penalty_cost)
    return expected_profit
```

### Part A

```python
revenues = {3:150000, 4:200000, 5:250000}
unit_cost = 20000
penalty_cost = 100000
lower_limit = 3
p_accepted = 0.65
expected_profits = []
for Q in range(1,11):
    expected_profits.append(calculate(Q, revenues, unit_cost, penalty_cost, lower_limit, p_accepted))
    print("Q = {:>2.0f}, profit = {:>10.2f}".format(Q,expected_profits[-1]))
print("------------------------------------")
print("Optimal quantity to manufacture is Q={:.0f} with profit {:.2f}".format(expected_profits.index(max(expected_profits))+1,max(expected_profits)))
```

```python
revenues = {3:150000, 4:200000, 5:250000}
unit_cost = 20000
penalty_cost = 100000
lower_limit = 3
expected_profits = {}
for p_accepted in np.arange(0.5, 0.96, 0.01):
    if p_accepted not in expected_profits.keys():
        expected_profits[p_accepted] = (-1,-float('inf'))
    for Q in range(1,11):
        expected_profit = calculate(Q, revenues, unit_cost, penalty_cost, lower_limit, p_accepted)
        if expected_profit > expected_profits[p_accepted][1]:
            expected_profits[p_accepted] = (Q, expected_profit)
fig, ax = plt.subplots(figsize=(7,5), dpi=100)
ax.plot(list(expected_profits.keys()), [i[0] for i in expected_profits.values()])
ax.set_xlabel("Acceptance probability")
ax.set_ylabel("Optimal production quantity")
```

###Â Part D

```python
def calculate_loss_risk(Q, revenues, unit_cost, penalty_cost, lower_limit, p_accepted):
    expected_prods = np.asarray([binom.pmf(k=i, n=Q, p=p_accepted) for i in range(Q+1)])
    total_revenues = np.asarray([revenues[i] if i in list(revenues.keys()) else revenues[max(list(revenues.keys()))] if i > max(list(revenues.keys())) else 0 for i in range(Q+1)])
    total_prod_cost = np.asarray([unit_cost*i for i in range(Q+1)])
    total_penalty_cost = np.asarray([penalty_cost if i<lower_limit else 0 for i in range(Q+1)])
    profits = total_revenues-total_prod_cost-total_penalty_cost
    loss_probs = np.clip(-profits / np.abs(profits), 0, 1)
    loss_probability = np.dot(expected_prods, loss_probs)
    #print(expected_prods)
    #print(loss_probs)
    return loss_probability
```

```python
revenues = {3:150000, 4:200000, 5:250000}
unit_cost = 20000
penalty_cost = 100000
lower_limit = 3
p_accepted = 0.65
loss_risks = []
for Q in range(1,11):
    loss_risks.append(calculate_loss_risk(Q, revenues, unit_cost, penalty_cost, lower_limit, p_accepted))
    print("Q = {:>2.0f}, loss probability = {:>4.2f}".format(Q,loss_risks[-1]))
```

##Â Question 2

```python
def calculate(Q, unit_sell_cost, unit_prod_cost, fixed_penalty_cost, penalty_cost_per_unit, lower_limit, p_accepted, recycle_value, salvage_value, demand=25):
    prod_probability = np.asarray([binom.pmf(k=i, n=Q, p=p_accepted) for i in range(Q+1)])
    total_revenues = np.asarray([unit_sell_cost*min(demand,i) for i in range(Q+1)])
    total_prod_cost = np.asarray([unit_prod_cost*Q for i in range(Q+1)])
    total_penalty_cost = np.asarray([penalty_cost+penalty_cost_per_unit*max(0,demand-i) if i<lower_limit else 0 for i in range(Q+1)])
    total_recycle_value = np.asarray([recycle_value*max(0,i-demand) for i in range(Q+1)])
    total_salvage_value = np.asarray([salvage_value*(Q-i) for i in range(Q+1)])
    expected_profit = np.dot(prod_probability, total_revenues+total_recycle_value+total_salvage_value-total_prod_cost-total_penalty_cost) 
    return expected_profit
```

```python
expected_profits = []
unit_sell_cost = 5000
unit_prod_cost = 2500
fixed_penalty_cost = 10000
penalty_cost_per_unit = 500
penalty_lower_limit = 25
p_accepted = 0.75
recycle_value = 900
salvage_value = 100
for Q in np.arange(0,101,1):
    expected_profits.append(calculate(Q, unit_sell_cost, unit_prod_cost, fixed_penalty_cost, penalty_cost_per_unit, penalty_lower_limit, p_accepted, recycle_value, salvage_value))
print("Optimal quantity to manufacture is Q={:.0f} with profit {:.2f}".format(expected_profits.index(max(expected_profits))+1,max(expected_profits)))
optimal_profit1 = max(expected_profits)
```
